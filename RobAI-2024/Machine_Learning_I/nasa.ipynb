# The ML Used Model (Sequential Deep Learning Model)

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=[len(x_train.keys())]),
    tf.keras.layers.Dense(units=256, activation="relu"),
    tf.keras.layers.Dense(units=256, activation="relu"),
    tf.keras.layers.Dense(units=256, activation="relu"),
    tf.keras.layers.Dense(units=256, activation="relu"),
    tf.keras.layers.Dense(units=256, activation="relu"),
    tf.keras.layers.Dense(units=1)
])
```

```python
def model_fit(model, optimizer_a, x, y, batch_size, epochs, validation_data):
    model.compile(loss="mse", optimizer=tf.keras.optimizers.RMSprop(optimizer_a), metrics=["mae", tf.keras.metrics.RootMeanSquaredError(), "accuracy"])
    model.fit(x=x, y=y, batch_size=batch_size, epochs=epochs, validation_data=validation_data, verbose=1)
```

# Train The Model Using Cross Validation (Four-Fold Cross Validation)

```python
df_size, tmp = df.shape
print(f"df size is {df_size}")

x_train1 = x_train[0 : (df_size*3)//4]
print(f"x_train1 shape {x_train1.shape}")
y_train1 = y_train[0 : (df_size*3)//4]
print(f"y_train1 shape {y_train1.shape}")

x_test1 = x_train[(df_size*3)//4 :]
print(f"x_test1 shape {x_test1.shape}")
y_test1 = y_train[(df_size*3)//4 :]
print(f"y_test2 shape {y_test1.shape}")
```

```python
model_fit(model=model, optimizer_a=0.002, x=x_train1, y=y_train1, batch_size=5, epochs=100, validation_data=(x_test1, y_test1))
```

```python
model_fit(model=model, optimizer_a=0.0001, x=x_train1, y=y_train1, batch_size=5, epochs=20, validation_data=(x_test1, y_test1))
```

```python
x_train2 = x_train.drop(x_train.index[(df_size*2)//4 : (df_size*3)//4])
print(f"x_train2 shape {x_train2.shape}")
y_train2 = y_train.drop(y_train.index[(df_size*2)//4 : (df_size*3)//4])
print(f"y_train2 shape {y_train2.shape}")

x_test2 = x_train[(df_size*2)//4 : (df_size*3)//4]
print(f"x_test2 shape {x_test2.shape}")
y_test2 = y_train[(df_size*2)//4 : (df_size*3)//4]
print(f"y_test2 shape {y_test2.shape}")
```
